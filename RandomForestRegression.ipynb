```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# import seaborn as sns
# from scipy import stats
from matplotlib import rc
# import missingno as msno

#엑셀파일 읽어오기
train_df = pd.read_csv('./data/data.csv')
train_df.head()

#Target 파일 읽어오기
train_df_target = pd.read_csv('./data/train_target.csv')
train_df_target.head()
```

```python
#테스트 데이터 추출
# x = train_df.drop(train_df.index[66452:])
# x_target = train_df_target.drop(train_df.index[66452:])
# y = train_df.drop(train_df.index[:66452])
# y_target = train_df_target.drop(train_df.index[:66452])
```

```python
x = train_df.iloc[:66452,:]
x_target = train_df_target.iloc[:66452,:]
y = train_df.iloc[66452:,:]
y_target = train_df_target.iloc[66452:,:]
```

```python
#set_index 설정하기
x = x.set_index(['td', 'code'])
x_target = x_target.set_index(['td', 'code'])
y = y.set_index(['td', 'code'])
y_target = y_target.set_index(['td', 'code'])

# #NaN fillna
x.fillna(x.mean(), inplace=True)
y.fillna(y.mean(), inplace=True)
y.head()
```

```python
>>> from sklearn.ensemble import RandomForestRegressor
>>> from sklearn.datasets import make_regression
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0,n_estimators=100)
```

```python
regr.fit(x, x_target)
```

```python
print(regr.feature_importances_)
```

```python
pred3 = regr.predict(y)
```

```python
submission2 = pd.DataFrame(pred3, columns = ['target'], index = y.index)
submission2.head()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_target.values, pred3)
```
